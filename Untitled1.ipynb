```python
import pandas as pd
from prophet import Prophet
import plotly.express as px

def run_forecasting_from_excel(file_path="/content/Online_Retail.xlsx"):
    try:
        # Đọc dữ liệu từ Excel
        df = pd.read_excel(file_path)

        # ✅ Kiểm tra tên cột có đúng không
        required_columns = ["InvoiceDate", "Quantity", "UnitPrice"]
        for col in required_columns:
            if col not in df.columns:
                raise ValueError(f"❌ Thiếu cột bắt buộc: {col}")

        # ✅ Tiền xử lý (KHÔNG đổi tên cột)
        df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")
        df = df.dropna(subset=["InvoiceDate", "Quantity", "UnitPrice"])
        df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]

        df["Sales"] = df["Quantity"] * df["UnitPrice"]
        df["Date"] = df["InvoiceDate"].dt.date

        # ✅ Tổng hợp doanh số theo ngày
        daily_sales = df.groupby("Date")["Sales"].sum().reset_index()
        daily_sales.columns = ["ds", "y"]

        if len(daily_sales) < 2:
            print("❌ Không đủ dữ liệu để dự đoán.")
            return

        # ✅ Huấn luyện Prophet
        model = Prophet()
        model.fit(daily_sales)

        # ✅ Dự báo 7 ngày tới
        future = model.make_future_dataframe(periods=7)
        forecast = model.predict(future)

        # ✅ Vẽ biểu đồ
        fig = px.line(
            forecast,
            x="ds",
            y="yhat",
            title="📈 Dự báo doanh số theo ngày (từ Excel)",
            labels={"ds": "Ngày", "yhat": "Doanh số dự đoán"},
            markers=True
        )
        fig.update_layout(template="plotly_white")
        fig.show()

    except Exception as e:
        print("❌ Lỗi:", e)
run_forecasting_from_excel()
```

```python
import pandas as pd
from prophet import Prophet
import plotly.graph_objects as go

def run_forecasting_and_compare(file_path="Online_Retail.xlsx"):
    try:
        df = pd.read_excel(file_path)
        df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")
        df = df.dropna(subset=["InvoiceDate", "Quantity", "UnitPrice"])
        df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]

        df["Sales"] = df["Quantity"] * df["UnitPrice"]
        df["Date"] = df["InvoiceDate"].dt.date

        daily_sales = df.groupby("Date")["Sales"].sum().reset_index()
        daily_sales.columns = ["ds", "y"]

        # ✅ Số ngày dự báo
        forecast_days = 20

        actual_last_days = daily_sales.tail(forecast_days)
        actual_last_days.to_excel("actual_last_20_days.xlsx", index=False)

        training_data = daily_sales.iloc[:-forecast_days]

        if len(training_data) < 2:
            print("❌ Không đủ dữ liệu để huấn luyện.")
            return

        model = Prophet()
        model.fit(training_data)

        future = model.make_future_dataframe(periods=forecast_days)
        forecast = model.predict(future)

        forecast_last_days = forecast[["ds", "yhat"]].tail(forecast_days)
        forecast_last_days["ds"] = forecast_last_days["ds"].dt.date

        merged = pd.merge(forecast_last_days, actual_last_days, on="ds", how="inner")
        merged.rename(columns={"yhat": "Forecast", "y": "Actual"}, inplace=True)

        fig = go.Figure()
        fig.add_trace(go.Scatter(x=merged["ds"], y=merged["Actual"], mode='lines+markers', name='Thực tế'))
        fig.add_trace(go.Scatter(x=merged["ds"], y=merged["Forecast"], mode='lines+markers', name='Dự báo'))
        fig.update_layout(
            title="📊 So sánh Dự báo vs Thực tế (20 ngày cuối)",
            xaxis_title="Ngày",
            yaxis_title="Doanh số",
            template="plotly_white"
        )
        fig.show()

        print("✅ Đã lưu file: actual_last_20_days.xlsx")
        print("🔍 Bảng so sánh:")
        print(merged)

    except Exception as e:
        print("❌ Lỗi:", e)

run_forecasting_and_compare()
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

def run_recommendation(file_path="/content/Online_Retail.xlsx", min_support=0.05, top_n_products=30):
    try:
        # Đọc dữ liệu
        df = pd.read_excel(file_path)
        df = df.dropna(subset=["InvoiceNo", "StockCode", "Description"])
        df = df[df["Quantity"] > 0]
        df["InvoiceNo"] = df["InvoiceNo"].astype(str)

        # ✅ Chỉ giữ lại top sản phẩm bán chạy nhất
        top_products = df["Description"].value_counts().nlargest(top_n_products).index.tolist()
        df = df[df["Description"].isin(top_products)]

        # ✅ Pivot bảng giỏ hàng
        basket = df.groupby(["InvoiceNo", "Description"])["Quantity"].sum().unstack().fillna(0)
        basket = basket.applymap(lambda x: 1 if x > 0 else 0)

        # ✅ Apriori
        frequent_itemsets = apriori(basket, min_support=min_support, use_colnames=True)

        if frequent_itemsets.empty:
            return {"message": "❌ Không tìm thấy luật kết hợp đủ phổ biến."}

        rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
        top_rules = rules.sort_values("lift", ascending=False).head(10)

        # ✅ Biến kết quả dễ đọc
        recommendations = []
        for _, row in top_rules.iterrows():
            recommendations.append({
                "Nếu mua": list(row["antecedents"]),
                "thì nên mua": list(row["consequents"]),
                "support": round(row["support"], 3),
                "confidence": round(row["confidence"], 3),
                "lift": round(row["lift"], 2)
            })

        return recommendations

    except Exception as e:
        return {"error": str(e)}
```

```python
run_recommendation(min_support=0.05, top_n_products=30)
```

```python
import plotly.express as px
import pandas as pd

def generate_recommendation_plot(rules_data: list):
    df = pd.DataFrame(rules_data)
    df["Nếu mua"] = df["Nếu mua"].apply(lambda x: ', '.join(x))
    df["thì nên mua"] = df["thì nên mua"].apply(lambda x: ', '.join(x))

    fig = px.scatter(
        df,
        x="support",
        y="confidence",
        size="lift",
        color="Nếu mua",
        hover_data=["Nếu mua", "thì nên mua", "lift"],
        title="📦 Đề xuất sản phẩm theo luật kết hợp",
    )
    fig.update_layout(template="plotly_white")
    return fig
```

```python
rules_data = run_recommendation(min_support=0.05, top_n_products=30)
fig = generate_recommendation_plot(rules_data)
fig.show()
```

```python
import pandas as pd
import json
import os
from prophet import Prophet

def run_forecasting():
    try:
        DATA_FILE = "/content/received_data.jsonl"

        if not os.path.exists(DATA_FILE):
            return {"error": "❌ Không tìm thấy file dữ liệu."}

        # 🧾 Load file
        data = []
        with open(DATA_FILE, "r", encoding="utf-8") as f:
            for line in f:
                try:
                    record = json.loads(line.strip())
                    data.append(record)
                except json.JSONDecodeError:
                    continue

        if not data:
            return {"error": "❌ File không có dữ liệu hợp lệ."}

        df = pd.DataFrame(data)
        print("✅ Số dòng dữ liệu đọc được:", len(df))

        # 📌 Kiểm tra cột bắt buộc
        required_cols = ["quantity", "unitPrice", "invoiceDate"]
        if not all(col in df.columns for col in required_cols):
            return {"error": f"❌ Thiếu các cột: {required_cols}"}

        # ⏳ Xử lý định dạng
        df["invoiceDate"] = pd.to_datetime(df["invoiceDate"], errors="coerce")
        df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")
        df["unitPrice"] = pd.to_numeric(df["unitPrice"], errors="coerce")

        df = df.dropna(subset=["quantity", "unitPrice", "invoiceDate"])
        df = df[(df["quantity"] > 0) & (df["unitPrice"] > 0)]
        df["sales"] = df["quantity"] * df["unitPrice"]

        print("✅ Sau xử lý còn:", len(df), "dòng")

        # 🕒 Tổng hợp theo phút
        df["minute"] = df["invoiceDate"].dt.floor("min")
        minute_sales = df.groupby("minute")["sales"].sum().reset_index()
        minute_sales.columns = ["ds", "y"]

        print("📊 Số phút có doanh số thực tế:", len(minute_sales))
        print("📅 Khoảng thời gian:", minute_sales['ds'].min(), "→", minute_sales['ds'].max())

        # 🔁 Điền các phút thiếu
        full_range = pd.date_range(start=minute_sales["ds"].min(), end=minute_sales["ds"].max(), freq="min")
        full_sales = pd.DataFrame({"ds": full_range})
        full_sales = full_sales.merge(minute_sales, on="ds", how="left").fillna(0)

        print("🔁 Tổng số phút sau khi điền đủ:", len(full_sales))
        print(full_sales.head())

        if len(full_sales) < 100:
            return {"error": "❌ Không đủ dữ liệu theo phút để dự đoán."}

        # 🤖 Dự báo
        model = Prophet()
        model.fit(full_sales)

        future = model.make_future_dataframe(periods=1440, freq="min")
        forecast = model.predict(future)

        result = [
            {
                "datetime": row["ds"].strftime("%Y-%m-%d %H:%M"),
                "predicted_sales": round(row["yhat"], 2)
            }
            for _, row in forecast.tail(1440).iterrows()
        ]

        print("✅ Dự đoán hoàn tất:", len(result), "phút tiếp theo")

        return result

    except Exception as e:
        return {"error": f"❌ Lỗi xử lý: {str(e)}"}
```

```python
import plotly.express as px

def draw_forecast_chart(data):
    df = pd.DataFrame(data)
    df["datetime"] = pd.to_datetime(df["datetime"])
    fig = px.line(
        df,
        x="datetime",
        y="predicted_sales",
        title="📈 Dự báo doanh số theo từng phút (24 giờ tiếp theo)",
        labels={"datetime": "Thời gian", "predicted_sales": "Doanh số dự đoán"},
        markers=True
    )
    fig.update_layout(template="plotly_white")
    fig.show()
```

```python
forecast_result = run_forecasting()
if isinstance(forecast_result, dict) and "error" in forecast_result:
    print(forecast_result["error"])
else:
    draw_forecast_chart(forecast_result)
```
